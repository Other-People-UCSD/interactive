```python
import json
import re

fileName = 'draft1'
openFile = './' + fileName + '.tsv'
writeFile = './' + fileName + '.json'
```

```python
with open(openFile, mode='r', encoding='utf8', newline='') as tsvFile:
    passages = []
    tsvFile.readline() # Headers
    for row in tsvFile:
        passageData = row.strip().split('\t')
        if (len(passageData) < 3):
            print('The data contains an improperly formatted line!')
            continue
        passageOpt = passageData[2:]
        # print('id[', passageData[0], ']:', len(re.split('  |\t', passageData[1])), passageData[1])
        print('id[', passageData[0], ']')

        keys = ['id', 'text', 'options']
        pOptKeys = ['to', 'text']

        pOptVals = []
        for i in range(0, len(passageOpt), 2):
            pOptVal = {
                'to': passageOpt[i],
                'text': passageOpt[i+1]
            }
            pOptVals.append(pOptVal)

        pid = passageData[0]

        
        pText = re.split('  |\t', passageData[1])

        passage = {
            keys[0]: pid,
            keys[1]: pText,
            keys[2]: pOptVals
        }
        passages.append(passage)

# print(passages)
jsonData = {
    'story': passages
}
jsonOut = json.dumps(jsonData)
```

```python
with open(writeFile, mode='w') as jsonFile:
    jsonFile.write(jsonOut)
```
